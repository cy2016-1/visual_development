# 伽马变换
伽马变换（幂律变换）是常用的灰度变换，是一种简单的图像增强算法：



我们可以得到：

- γ值小于1时，会拉伸图像中灰度级较低的区域，同时会压缩灰度级较高的部分
  
- γ值大于1时，会拉伸图像中灰度级较高的区域，同时会压缩灰度级较低的部分部分

输入数据r的取值范围为[0,1],所以我们需要对数据归一化

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('./exp_images/tsukuba.png',0).astype(np.float32)/255
gamma = 0.5
corrected = np.power(img, gamma)
cv.imshow("image",corrected)
cv.waitKey(0)
cv.destroyAllWindows()
```

变换γ值查看不同校正结果：

```python
gamma = [0.125, 0.25, 0.5, 1.0, 2.0, 4.0]
for k in range(len(gamma)):
    corrected = np.power(img,gamma[k])
    plt.subplot(2,3,k+1)
    plt.imshow(corrected,cmap='gray')
    plt.title(f"$\gamma={gamma[k]}$")
    plt.xticks([]);plt.yticks([])
plt.show()
```
